```python
from sympy import symbols, cos, dotprint
from graphviz import Source# for rendering
```

```python
a, b, c, x= symbols('a b c x')
expr = a * b * cos( c * x ) + b

src = Source(dotprint(expr))
src
#src.render('output.gv', view=True)
```

```python
#############  Convert textual representation to tree-representation (in sympy) 
from sympy import *

x0, x1, x2, x3, x4 = symbols('x0 x1 x2 x3 x4')
expression = "x2**(1/4)*log(2)*log(x0)/sqrt(x1**2)"
fs = sympify(expression)
sx = Source(dotprint(fs))
print(fs.count_ops())
print(len(fs.atoms()))
sx
```
